```python
from collections import defaultdict
 
class Graph:
    #Hàm tìm đỉnh với giá trị nhỏ nhất của những đỉnh chưa được đưa vào đường đi ngắn nhất
    def minDistance(self,dist,queue):
        # Initialize min value and min_index as -1
        minimum = float("Inf")
        min_index = -1
         
        #Từ mảng dist chọn ra đỉnh có chi phí thấp nhất
        for i in range(len(dist)):
            if dist[i] < minimum and i in queue:
                minimum = dist[i]
                min_index = i
        return min_index
 
 
    #Hàm in ra đường đi ngắn nhất
    def printPath(self, parent, j):
         
        
        #Trường hợp j là đỉnh bắt đầu
        if parent[j] == -1 :
            print(j,end=" ")
            return
        self.printPath(parent , parent[j])
        print (j,end=" ")
         
 
    #Hàm in khoảng cách từ đỉnh nguồn đến đỉnh hiện tại đang xét
    def printSolution(self, dist, parent):
        src = 0
        print("Vertex \t\tDistance from Source\tPath")
        for i in range(1, len(dist)):
            print("\n%d --> %d \t\t%d \t\t\t\t\t" % (src, i, dist[i]),end=" ")
            self.printPath(parent,i)
 

    def dijkstra(self, graph, src):
 
        row = len(graph)
        col = len(graph[0])
 
        #Tạo mảng dist chứa khoảng cách từ đỉnh nguồn đến đỉnh i, phẩn tử khởi tạo giá trị là vô cực
        dist = [float("Inf")] * row
 
        #Mảng parent lưu đường đi ngắn nhất
        parent = [-1] * row
 
        #khoảng cách của đỉnh nguồn đến chính nó = 0
        dist[src] = 0
     
        #Thêm tất cả các đỉnh vào trong hàng đợi
        queue = []
        for i in range(row):
            queue.append(i)
             
        #Tìm đường đi ngắn nhất trong tất cả các đỉnh
        while queue:
 
            #Tìm đỉnh có chi phí nhỏ nhất từ dist và vẫn còn trong hàng đợi
            u = self.minDistance(dist,queue)
 
            #Xóa đỉnh đã thêm vào đường đi ngắn nhất 
            queue.remove(u)
 
            #Cập nhât lại dist, các đỉnh liền kề với đỉnh đã chọn(chỉ xét đỉnh còn trong hàng đợi)
            for i in range(col):
                if graph[u][i] and i in queue:
                    if dist[u] + graph[u][i] < dist[i]:
                        dist[i] = dist[u] + graph[u][i]
                        parent[i] = u
 
 
        # in khoảng cách
        self.printSolution(dist,parent)
 
g= Graph()
 
graph = [[0, 4, 0, 0, 0, 0, 7, 0, 0],
        [4, 0, 9, 0, 0, 0, 11, 20, 0],
        [0, 9, 0, 6, 2, 0, 0, 0, 0],
        [0, 0, 6, 0, 10, 5, 0, 0, 0],
        [0, 0, 2, 10, 0, 15, 0, 1, 5],
        [0, 0, 0, 5, 15, 0, 0, 0, 12],
        [7, 11, 0, 0, 0, 0, 0, 1, 0],
        [0, 20, 0, 0, 1, 0, 1, 0, 3],
        [0, 0, 0, 0, 5, 12, 0, 3, 0]
        ];
 
g.dijkstra(graph,0)
```
